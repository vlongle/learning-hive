```python
from shell.fleet.network import TopologyGenerator
import matplotlib.pyplot as plt
```

```python
g = TopologyGenerator(num_nodes=8, edge_drop_prob=0.6)
```

```python
N = 12
fig, ax = plt.subplots(2, N//2, figsize=(5 * N//2, N))
ax = ax.flatten()
TopologyGenerator.plot_graph(g.generate_tree(), ax=ax[0])
TopologyGenerator.plot_graph(g.generate_server(), ax=ax[1])
TopologyGenerator.plot_graph(g.generate_ring(), ax=ax[2])
TopologyGenerator.plot_graph(g.generate_fully_connected(), ax=ax[3])
TopologyGenerator.plot_graph(g.generate_connected_random(edge_drop_prob=0.25), ax=ax[4])
TopologyGenerator.plot_graph(g.generate_connected_random(edge_drop_prob=0.5), ax=ax[5])
TopologyGenerator.plot_graph(g.generate_connected_random(edge_drop_prob=0.7), ax=ax[6])
TopologyGenerator.plot_graph(g.generate_connected_random(edge_drop_prob=0.9), ax=ax[7])

TopologyGenerator.plot_graph(g.generate_random(edge_drop_prob=0.25), ax=ax[8])
TopologyGenerator.plot_graph(g.generate_random(edge_drop_prob=0.5), ax=ax[9])
TopologyGenerator.plot_graph(g.generate_random(edge_drop_prob=0.7), ax=ax[10])
TopologyGenerator.plot_graph(g.generate_random(edge_drop_prob=0.9), ax=ax[11])
```
